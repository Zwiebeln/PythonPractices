```python
import numpy as np
from matplotlib import pyplot as plt
import time
```

```python
%matplotlib inline
```

```python
x= np.arange(100)
y = x+ np.random.normal(scale=5, size=100)
plt.plot(x,y,'.')
```

```python
def gradientW(w, x, b, y):
    return np.sum((w*x+b -y)*x)
```

```python
def gradientB(w, x, b, y):
    return np.sum(w*x+b-y)
```

```python
def cost(w,x,b,y):
    return (np.sum((w*x+b-y)**2))/len(x)*2
```

```python
#initiate Value
def initValue(x,y):
    w = np.random.random()
    b = np.random.random()
    gw = gradientW(w,x,b,y)
    gb = gradientB(w,x,b,y)
    print("Initial w and b: w: %f b: %f" %(w,b))
    return w,b,gw,gb
```

```python
no_ite_visual=[]
cost_visual =[]
def gradientDescent(x, y, epsilon):
    startTime= time.clock()
    learningRate = 0.01
    w,b,gw,gb= initValue(x,y)
    prev_cost = cost(w,x,b,y)
    print("prev_cost %f" %prev_cost)
    no_iteration = 0
    while abs(gw) > epsilon and abs(gb) > epsilon: #终止条件
        no_iteration +=1
        curr_cost = cost(w,x,b,y)
        if curr_cost > prev_cost:
            learningRate /= 2.0
            prev_cost = curr_cost
            w = prev_w
            b = prev_b
        gw = gradientW(w,x,b,y)
        gb = gradientB(w,x,b,y)
        prev_w = w
        prev_b = b
        w -= learningRate*gw
        b -= learningRate*gb
        if(no_iteration%1000==0):
            no_ite_visual.append(no_iteration)
            cost_visual.append(curr_cost)
        print(curr_cost)
        #print ("w: %f, b: %f" %(w,b))
    print( no_iteration)
    stoptime= time.clock()
    duration= stoptime- startTime
    print("Final w and b: w: %f b: %f" %(w,b))
    print ("Duration: %f s" %duration)
```

```python
gradientDescent(x,y,1e-1)
```
