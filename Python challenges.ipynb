```python
def rearrange(st):
    res=''
    for i in st:
        if (ord(i)>=ord('a') and ord(i)<=ord('z')):
            res+= chr((ord(i)-ord('a')+2)%26+ord('a'))
        else:
            res+=i
    return res
```
